```python
import pandas as pd
import numpy as np 

df=pd.read_csv('fifa21_raw_data.csv', low_memory=False)
df.head()
```

```python
# converting the height column from string to numeric and splitting it into feet and inches
df['Height']=df['Height'].str.replace(r'[^0-9]+',' ',regex=True)
df[['height_feet','height_inch']]=df['Height'].str.split(expand=True)
df[['height_feet', 'height_inch']] = df[['height_feet', 'height_inch']].apply(pd.to_numeric, errors='coerce')
```

```python
# Converting feet to total inches and centimetres:

df['height_in_inches']= (df['height_feet']*12)+df['height_inch']
df['height_in_centimetres']=df['height_in_inches']*2.54
df
```

```python
#cleaning weight column and converting to numeric

df['weight_in_pounds']=df['Weight'].str.replace(r'[^0-9]',' ',regex=True)
df['weight_in_pounds']= df['weight_in_pounds'].apply(pd.to_numeric, errors='coerce')
df['weight_in_pounds']
```

```python
# converting string to datetime
from datetime import datetime
df['Joined_new']= pd.to_datetime(df['Joined'], format='%b %d, %Y')
df['Joined_new']
```

```python
# copying to new dataframe
select=['Joined_new','weight_in_pounds']
df2=pd.DataFrame(columns=select)
df2=df[select].copy()
df2
```

```python
#Cleaning the Value column and converting it into numeric type
df['Value']=df['Value'].replace(' ','')

df2[['Currency','Value','Multiplier']]=df['Value'].str.extract(r'([^\d.]*)([0-9.]+)([a-zA-Z]*)')

#df2=df2.drop(df2.columns[2], axis=1)

df2['Multiplier'] = df2['Multiplier'].replace({'M': 1000000, 'm': 1000000}, regex=True)
df2['Multiplier'] = df2['Multiplier'].replace({'K': 1000, 'k': 1000}, regex=True)

df2['Value']=df2['Value'].apply(pd.to_numeric, errors='coerce')
df2['Multiplier']=df2['Multiplier'].apply(pd.to_numeric, errors='coerce')

df2['Total_value']= df2['Value']*df2['Multiplier']
#df2['Total_value']=df2['Total_value'].round()
```
